# **Pseudoalgoritmos**

Un pseudoalgoritmo es una forma de representar cualquier algoritmo sin necesidad de incluir un lenguaje de programación en concreto. Se caracteríza por no ser realmente ejecutable, sino que representa una una idea lógica para ser transmitida arbitrariamentre. El programador es libre de llevar a cabo esta idea lógica en cualquier lenguaje de programación y ejecutarla de una manera personalizada. Un ejemplo de pseudoalgoritmo sería el siguiente:

1.jpg

El pseudoalgoritmo generaliza la idea del famoso algoritmo *Insertion sort*, el cual inicia indicando la información que necesita de entrada y la que arroja como resultado, que en este caso sería el array desordenado y el ordenado, respectivamente. Luego, si se quisiera implementar el anterior pseudoalgoritmo en Python, sería de la forma:

```python
def InsertionSort(A):
  N = len(A)
  for j in range(2,N):
    key = A[j]
    i = j-1
    while (i>=0) and (A[i]>key):
      A[i+1]=A[i]
      i = i-1
    A[i+1] = key
  return A

A = [1, 9, 2, 7, 1, 6, 0, 4]
B = InsertionSort(A)
print(B)
```

Los pseudoalgoritmos son fundamentales para transmitir ideas algorítmicas sin necesidad de resringirlas a un lenguaje de programación en concreto. Estos son muy usados en textos de optimización e inteligencia artificial debido a su versatilidad, por lo que la idea de este laboratorio es familiarizarse con ellos y, posteriormente, ejecutarlos.

 El objetivo de este algoritmo es generar la trayectoria de una variable $x$ durante $n$ pasos, donde el valor de la variable en el siguiente paso depende del valor de la variable en el paso actual. Este es el tipo de algoritmos usados en inteligencia artificial para guiar los procesos de aprendizaje de las máquinas. En este ejemplo, tenemos un proceso iterativo de $n$  pasos que actualiza un arreglo $x$. El algoritmo recibe como parámetros de entrada una condición inicial $x[0]$ y el número de pasos $n$, y debe retornar el arreglo final $x$.

**2.1.** Implemente el algoritmo anterior como una función y realice una prueba de ejemplo.  

```python
#Ejecución: implemente y ejecute su algoritmo aquí
import math

# El algorithmo es llamado eml (encontrar minimo local)
def eml (x:int,n:int ):
  d =[]
  x_array =[]
  x_array.append(x)
  for k in range(0,n-1):
    x_k = x_array[k]
    d.append(round(math.cos(x_k)+((10/3)* math.cos((10/3)*x_k)),2))
    x_array.append(round(x_k-(0.1*d[k]),2))
  return x_array
eml(3,20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.sin(x) + np.sin((10/3) * x)
x = np.linspace(2.7, 7.5,100)
y = f(x)
x_k = [3]
n = 100
for k in range(0, n):
    x_next = x_k[0] +  f(x_k[0])*-1
    x_k.append(x_next)
x_k = np.array(x_k)
y_k = f(x_k)
plt.plot(x, y, label='Función', color='blue')
plt.scatter(x_k,y_k,color='orange', label='Trayectoria')
plt.plot(x_k, y_k, linestyle='solid', color='orange')
plt.xlabel('x[k]')
plt.ylabel('f(x[k])')
plt.legend()
plt.title("Trayectoria")
```
